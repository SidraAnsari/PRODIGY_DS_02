```python
#Data Cleaning performed
dataset['Age'].fillna(dataset['Age'].median(),inplace=True)
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0],inplace=True)
```

```python
dataset['Cabin'].fillna(dataset['Cabin'].mode()[0],inplace=True)
```

# Exploratory Data Analysis(EDA) Performing here

```python
#Relationship between Variables: Survival and Age
plt.figure(figsize=(12,8))
sns.histplot(data=dataset,x='Age',hue='Survived',kde=True)
plt.title('Survival w.r.t Age')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data=dataset,x="Survived",y="Sex")
plt.title("Survival w.r.t Sex")
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=dataset,x="Survived",hue="Sex")
plt.title("Survival w.r.t Sex")
plt.show()
```

```python
# Survival and Embarked
sns.set(style="whitegrid")
plt.figure(figsize=(10, 4))
sns.countplot(data=dataset,x="Survived", hue="Embarked")
plt.title("Survival w.r.t Embarked")
plt.show()
```

```python
# Survival vs. fare
plt.figure(figsize=(10, 4))
sns.histplot(data=dataset, x='Fare', kde=True, hue='Survived')
plt.title("Survival w.r.t Fare")
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.scatterplot(data=dataset, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
